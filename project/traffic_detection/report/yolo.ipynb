```python
from ultralytics import YOLO
```

# YOLO 8

## Chưa enhance

```python
# Load mô hình YOLOv8
model = YOLO('yolov8s.pt') 
```

```python
# Huấn luyện mô hình với early stopping
model.train(
    data='/kaggle/input/hphat-yaml/kaggle.yaml',  # Tệp cấu hình dữ liệu
    epochs=40,                            # Số epoch tối đa
    imgsz=640,                            # Kích thước ảnh đầu vào
    batch=32,                             # Kích thước batch
    patience=5                            # Dừng sớm sau 5 epoch nếu không cải thiện
)
```

```python
model.save('yolov8s_vehicle.pt')
```
